## Import necessary packages

```python
# For numerical array
import numpy as np
# For plotting:
import matplotlib.pyplot as plt
# To ensure that the plots actually do appear in this notebook:
% matplotlib inline
```

## Plotting curves

```python
# Just create a linearly spaced 1D array
time_vec = np.linspace(0, 2 * np.pi, num=20)
print('Shape of time_vec: {}'.format(time_vec.shape))

# Now plot this array:
plt.plot(time_vec)
```

### Plotting a curve against a reference:

```python
sine_vec = np.sin(time_vec)
plt.plot(time_vec, sine_vec)
```

### Plenty of customization options

```python
plt.plot(time_vec, sine_vec,
         color='green', 
         marker='s', 
         linestyle='dashed',
         linewidth=2, 
         markersize=10)

plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (V)', fontsize=16)
plt.title('My awesome signal', fontsize=18);
```

## Scatter plot:

```python
reference_vec = np.arange(16)
observations = reference_vec ** 3
plt.scatter(reference_vec, observations)
```

## Bar plot:

```python
plt.bar(reference_vec, observations)
plt.xlabel('Year')
plt.ylabel('Profits')
plt.title('My fake business')
```

## 2D Images

```python
x_vec = np.linspace(0, 2*np.pi, 256)
y_vec = np.sin(x_vec)
two_d_image = y_vec * np.atleast_2d(y_vec).T

print('Shape of the matrix: {}'.format(two_d_image.shape))

plt.imshow(two_d_image);
```

### Plenty of customization options again

```python
plt.imshow(two_d_image, cmap=plt.cm.jet, origin='lower', vmin=-0.5, vmax=0.5)
```

## The recommended way to use matplotlib:

``plt.imshow``, ``plt.plot`` and the other functions will most certainly be quick and easy to use but are **not the recommend** route. 

The recommended method is to explicitely create the figure and axes. This will not be covered here but 

```python
fig, axis = plt.subplots()
axis.imshow(two_d_image);
```
